```python
import pandas as pd
import io
import requests
url = "https://sdmx.data.unicef.org/ws/public/sdmxapi/rest/data/UNICEF,CME_DF_2020a,1.0/AFG+BRA+CHN+ISR+RUS+TUR+USA+VEN.MRY0T4+MRY1T4+PR_MRY0T4+TMY5T24+TMY5T9._T+F+M..?format=csv&labels=both"
data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv)
```
